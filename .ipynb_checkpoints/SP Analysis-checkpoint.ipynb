### Import libraries

```r
#ggplot2 for visualization
#readr for csv or readxl for excel
library(ggplot2)
library(readr)
#library(readxl)
```

### Read CSV

```r
sprite <- read_csv("data-10.csv")
```

### Select by year

```r
sprite_2017<-sprite[sprite['Año']==2017,]
sprite_2016<-sprite[sprite['Año']==2016,]
```

### Boxplot by month

```r
#ggplot(data,aes(x_axis,y_axis))
#geom_boxplot for boxplotting
    #groupped by 'Mes'
#scale_x and scale_y is to define x or y axis scales
#labs for labels in x or y axi, and main ttile
ggplot(sprite_2017, aes(Mes, `Sum of Activity Total`)) + 
geom_boxplot(aes(group= cut_width(Mes,0.25)), outlier.colour = "red", outlier.shape = 1, colour = 'tomato3') + 
scale_x_continuous(breaks = seq(min(sprite_2017$Mes),max(sprite_2017$Mes),by=1))+ 
scale_y_continuous(breaks = seq(0,max(sprite_2017$`Sum of Activity Total`),by=500))+
labs(x ='Months', y='Activity', title='Activity by month 2017')
```

```r
ggplot(sprite_2016, aes(Mes, `Sum of Activity Total`)) + 
geom_boxplot(aes(group= cut_width(Mes,0.25)), outlier.colour = "red", outlier.shape = 1, colour = 'tomato3') + 
scale_x_continuous(breaks = seq(min(sprite_2016$Mes),max(sprite_2016$Mes),by=1))+
scale_y_continuous(breaks = seq(0,max(sprite_2016$`Sum of Activity Total`),by=500))+
labs(x ='Months', y='Activity', title='Activity by month 2016')
```
